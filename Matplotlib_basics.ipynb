## Line Charts

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# create soem dummy data
#linspace creates one dimensional data with 1000 data points starting from 0 at a space of 20 points each
x= np.linspace(0,20,1000)
```

```python
# generate some artitary y-axis data
y= np.sin(x) + 0.2*x
```

```python
# plot 
plt.plot(x,y)
```

```python
# if not using a jupyter notebook then use show() method to print the graph
plt.plot(x,y)
plt.show()
```

```python
# to add labels and title
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
# to avoid showing the output of title, use semicolon
plt.title('Line plot');
```

```python
# import Pillow library
from PIL import Image
```

```python
im = Image.open('Images/Q2_kendall_1.jpg')
```

```python
#plot the image
plt.imshow(im);
```

```python
type(im)
```

```python
# convert image to a numpy array
arrImg = np.array(im)
```

```python
#shape of this array - three dimensional
# first two are spatial - height and width
# third dimensional - color - RGB - how much Red, green and blue to combine to make this images
arrImg.shape
```

```python
# print the array
arrImg
```

```python
# numbers are between 0 to 255
```

```python
# can also plot using the array
plt.imshow(arrImg);
```

```python
# to convert an image into a grey channel - black and white
# take the mean 
gray = arrImg.mean(axis=2)
```

```python
# check the shape - the third dimension should go away since its only black and white now , no RGB
gray.shape
```

```python
# plot the gray image
plt.imshow(gray);
```

```python
# to plot in gray scale
plt.imshow(gray, cmap='gray')
```
