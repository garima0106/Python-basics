```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

## Scipy to find pdf , cdf

Scipy is a library like numpy but has lot of functions to do statistical computations, signal processing, computer vision etc.

```python
x= np.linspace(-5,5, 1000)
```

```python
# find the probability distribution function (pdf) of x
fx = norm.pdf(x, loc=0, scale=1)
```

```python
plt.plot(x,fx);
```

```python
# find the cdf
Fx = norm.cdf(x, loc=0, scale=1)
```

```python
plt.plot(x, Fx);
```

```python
# logpdf
logfx= norm.logpdf(x, loc=0, scale=1)
```

```python
plt.plot(x, logfx);
```

## Convolution

```python
from PIL import Image
```

```python
#open the image
im= Image.open('Images/Q2_kendall_1.jpg')
```

```python
gray= np.mean(im, axis=2)
```

```python
plt.imshow(gray, cmap='gray');
```

```python
# use convolution to filter the image
```

```python
# create 2-dimensional gaussian filter
x= np.linspace(-6,6,50)
# find pdf for these points
fx= norm.pdf(x, loc=0, scale=1)
```

```python
# gaussian filter is a 2-dimensional spherical constants with some constants
# to calculate this gaussian filter we can do a outer product of pdf 
filt = np.outer(fx,fx)
```

```python
# plot the filter
plt.imshow(filt, cmap='gray');
```

```python
from scipy.signal import convolve2d
```

```python
# now apply the filter to the image using convolution
out=convolve2d(gray, filt)
```

```python
# now lets plot both the figures
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(out, cmap='gray');
```

## Edge Detection

```python
# define filters
hx = [[1,0,-1], [2,0,-2], [1,0,-1]]
hy= [[1,2,1], [0,0,0],[-1,-2,-1]]
```

```python
# apply the filters to the image and take convolution
gx= convolve2d(gray,hx)
gy= convolve2d(gray,hy)
```

```python
# for edge detection
edge=np.sqrt(np.square(gx)+np.square(gy))
```

```python
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(edge, cmap='gray');
```
